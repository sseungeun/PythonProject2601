# 소수판별 is_prime()
[2,3,4,5,6,7,8,9] -> [true, true, false, true, false, true, false, false]

```python
def is_prime(n):
    result = []

    for x in n:                 # x가 리스트 안의 숫자 하나
        if x < 2:
            result.append(False)
            continue

        isPrime = True
        for d in range(2, x):   # 2부터 x-1까지
            if x % d == 0:
                isPrime = False
                break

        result.append(isPrime)

    return result

print(is_prime([2,3,4,5,6,7,8,9]))
# [True, True, False, True, False, True, False, False]

```

# 등수 구하기 - 평균점수(영어점수, 수학점수)

```python
score1=[[80,70],[90,50],[40,70],[50,80]]

avgs=[]
for scores in score1:
    avgs.append(sum(scores)/len(scores))

ranks=[]
for a in avgs:
    rank = 0
    for b in avgs:
        if b>a:
            rank+=1
            ranks.append(rank)

for i in range(len(score1)):
    print(f"{i+1}번: 점수={score1[i]}, 평균={avgs[i]:.2f}, 등수={ranks[i]}")
# rank(score1) # [1,2,4,3]
```
